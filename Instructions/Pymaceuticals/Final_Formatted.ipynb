## Correlation and Regression

```python
# Calculate the correlation coefficient
# for mouse weight and average tumor volume for the Capomulin regimen
correlation = st.pearsonr(mouse_weight,avg_tumor_vol)
print(f"The correlation between both factors is {round(correlation[0],2)}")
```

```python
#Linear regression for mouse weight and average tumor volume for the Capomulin regimen
mwv_slope, mwv_int, mwv_r, mwv_p, mwv_std_err = st.linregress(mouse_weight, avg_tumor_vol)
```

```python
# Create equation of line to calculate predicted mouse weight
mwv_fit = mwv_slope * mouse_weight + mwv_int
```

```python
# Plot the linear regression model on top of the previous scatter plot.
mouse_weight_x = capomulin_values.iloc[:, 4]
avg_tumor_vol_y = capomulin_values.iloc[:, 6]
(mwv_slope, mwv_intercept, mwv_rvalue, mwv_pvalue, mwv_stderr) = linregress(mouse_weight_x, avg_tumor_vol_y)
regress_values = mouse_weight_x * mwv_slope + mwv_intercept
line_eq = "y = " + str(round(mwv_slope,2)) + "x + " + str(round(mwv_intercept,2))
plt.scatter(mouse_weight_x,avg_tumor_vol_y)
plt.plot(mouse_weight_x,mwv_fit,"--")
plt.annotate(line_eq,(6,10),fontsize=15,color="blue")
plt.xticks(mouse_weight, rotation=90)
plt.xlabel('Mouse Weight')
plt.ylabel('Tumor Volume')
plt.title("Linear Regression Model")
plt.show() 
```
